```python
import open3d as o3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
color_raw = o3d.io.read_image("image/framergb11.png")
depth_raw = o3d.io.read_image("depth/framedepth11.png")
rgbd_image = o3d.geometry.RGBDImage.create_from_color_and_depth(
    color_raw, depth_raw)
print(rgbd_image)
```

```python
plt.subplot(1, 2, 1)
plt.title('image')
plt.imshow(rgbd_image.color)
plt.subplot(1, 2, 2)
plt.title('depth image')
plt.imshow(rgbd_image.depth)
plt.show()
```
